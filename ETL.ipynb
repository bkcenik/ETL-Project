## import dependencies

```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

## read in and clean data

```python
education_2014_df = pd.read_csv("2014.csv")
education_2014_df.head()
```

```python
new_education_2014_df = pd.DataFrame({
    "State": education_2014_df["STATE"],
    "Enrollment": education_2014_df["ENROLL"],
    "Revenue": education_2014_df["TOTAL_REVENUE"],
    "Expenditure": education_2014_df["TOTAL_EXPENDITURE"]
})
new_education_2014_df.set_index("State")
new_education_2014_df.head()
```

```python
education_2013_df = pd.read_csv("2013.csv")
education_2013_df.head()
```

```python
new_education_2013_df = pd.DataFrame({
    "State": education_2013_df["STATE"],
    "Enrollment": education_2013_df["ENROLL"],
    "Revenue": education_2013_df["TOTAL_REVENUE"],
    "Expenditure": education_2013_df["TOTAL_EXPENDITURE"]
})
new_education_2013_df.set_index("State")
new_education_2013_df.head()
```

```python
crime_df = pd.read_csv("SentencingProject.csv")
crime_df.head()
crime_df["State"].str.upper()
crime_df.head()
```

```python
new_crime_df = pd.DataFrame({
    "State": crime_df["State"],
    "prison_pop": crime_df["PrisonPop"],
    "prison_rate": crime_df["PrisonRate100k"],
    "jail_pop": crime_df["JailPop"],
    "jail_rate": crime_df["JailRate100k"]
})
new_crime_df.head()
```

```python
census_df = pd.read_csv("census.csv")
census_df.head()
```

```python
new_census_df = pd.DataFrame({
    "State": census_df["State"],
    "Population": census_df["2013"]
})
new_census_df.head()
```

```python
# state, enrollment, revenue, expenditure
# state, prison_pop, prison_rate, jail_pop, jail_rate
# state, population
```

## create database connection

```python
connection_string = f'mysql://root:{password}@localhost:3306/etl_project_db'
engine = create_engine(connection_string)
```

```python
engine.table_names()
```

## load dataframes into database

```python
new_education_2013_df.to_sql(name='education', con=engine, if_exists='append', index=False)
```

```python
new_crime_df.to_sql(name='crime', con=engine, if_exists='append', index=False)
```

```python
new_census_df.to_sql(name='population', con=engine, if_exists='append', index=False)
```

## SQL queries to check data has been loaded

```python
pd.read_sql_query('select * from education', con=engine).head()
```

```python
pd.read_sql_query('select * from crime', con=engine).head()
```
